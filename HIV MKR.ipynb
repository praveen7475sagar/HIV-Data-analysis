```python
import pandas as pd
```

```python
#1st data set
df1=pd.read_csv("C:\\Users\\ARUN PRAVEEN\\Desktop\\HIV sheet1.csv")
```

```python
df1.head()
```

```python
for i in df1.columns:
    if i!='Year':
        df1[i]=df1[i].str.replace(r'\[.*\]',"")
```

```python
df1.head()
```

```python
Length=len(df1)
for i in range(Length):
    s=df1['Number of new HIV infections'][i].split(" ")
    sol=" "
    for j in range(len(s)):
        sol+=s[j]
    df1['Number of new HIV infections'][i]=int(sol)
```
